```python
import pandas as pd

df = pd.read_csv("Log of complaints.csv", encoding='windows-1252')

df.head()
df.info()
df.isnull().sum()
```

```python
# 1. Convert date
df['created_at'] = pd.to_datetime(df['created_at'], format='mixed')

# 2. Drop useless columns
df = df.drop(columns=[
    'title', 'description',
    'sub_category_id', 'category_id',
    'civic_agency_id'
])

# 3. Handle missing values

# Drop rows where critical info missing
df = df.dropna(subset=['ward_title', 'category_title'])

# Fill less important ones
df['civic_agency_title'] = df['civic_agency_title'].fillna('Unknown')

# Drop address (too many missing)
df = df.drop(columns=['address'])

df.head()
```

```python
df.shape
```

```python
df['category_title'].value_counts()
```

```python
df['category_title'] = df['category_title'].replace({
    'Streetlights': 'Street lighting',
    'Roads and Footpaths': 'Mobility - Roads, Footpaths and Infrastructure',
    'Mobility - Roads, Public transport': 'Mobility - Roads, Footpaths and Infrastructure',
    'Water Supply': 'Water Supply and Services',
    'Electricity & Power': 'Electricity and Power Supply',
    'Power supply': 'Electricity and Power Supply',
    'Solid Waste Management': 'Garbage and Unsanitary Practices',
    'Waste Management': 'Garbage and Unsanitary Practices'
})
```

```python
df['category_title'].value_counts()
```

```python
df['year'] = df['created_at'].dt.year

df['year'].value_counts().sort_index()
```

```python
df['month'] = df['created_at'].dt.to_period('M')

df['month'].value_counts().sort_index()
```

```python
df['ward_title'].value_counts().head(10)
```

```python
top_wards = df['ward_title'].value_counts().head(5).index

df[df['ward_title'].isin(top_wards)] \
    .groupby(['ward_title', 'category_title']) \
    .size() \
    .sort_values(ascending=False)
```

```python
top_wards = df[df['ward_title'] != 'Other']['ward_title'].value_counts().head(5).index

df[df['ward_title'].isin(top_wards)] \
    .groupby(['ward_title', 'category_title']) \
    .size() \
    .reset_index(name='count') \
    .sort_values(['ward_title', 'count'], ascending=[True, False])
```
